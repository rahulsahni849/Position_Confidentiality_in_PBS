```python
from matplotlib import pyplot as plt
import datetime
import pandas as pd
```

```python
def PerformanceCalc(queries,CTs):
    result = CTs/queries
    return round(result,4)


def performance_test():
    df = pd.read_csv('./anlysis_data/performance_data')
    pbs_queries = df.iloc[:,0]
    avg_cloacking_time  = df.iloc[:,1]

    WSSCT = []
    for i in range(0,df.shape[0]):
        #print(pbs_queries[i],avg_cloacking_time[i])
        WSSCT.append(PerformanceCalc(pbs_queries[i],avg_cloacking_time[i]))
        
    # print(pbs_queries)
    # print(WSSCT)
    plt.plot(pbs_queries,WSSCT)

    #creating Dataframe 
    d1 = pd.DataFrame(list(zip(pbs_queries,WSSCT)),columns=["PTS","Proposed-Method"])
    print(d1)

    plt.xlabel("Number of Position Track Sequence")
    plt.ylabel("Wireless Search Space Cloaking Region ( WSSCR )")
    plt.title("Performance Test")
    plt.show()

performance_test()

```

```python

from turtle import color


def PCA_Calc(Cachived,CLOACKS):
    result = round(Cachived/CLOACKS,2)
    return result
    
def PCA_test():
    df = pd.read_csv('./anlysis_data/PCA_testdata')
    PTS = df.iloc[:,0]
    CLOACKS  = df.iloc[:,1]

    PCA = []
    for i in range(0,df.shape[0]):
        #print(pbs_queries[i],avg_cloacking_time[i])
        PCA.append(PCA_Calc(PTS[i],CLOACKS[i]))
        
    plt.plot(PTS,PCA,color="red",animated=True)

    #creating Dataframe 
    d1 = pd.DataFrame(list(zip(PTS,PCA)),columns=["PTS","Proposed-Method"])
    print(d1)

    plt.xlabel("Number of Position Track Sequence")
    plt.ylabel("Position Confidentiality Assurance ( % )")
    plt.title("PCA Test")
    plt.show()



PCA_test()
```
